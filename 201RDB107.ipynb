```python
 # ⚠️⚠️⚠️⚠️ Circles
circles = []
diametr = 20

class cicrle:    
    def __init__(self, x, y, name, shouldAppend = True):
        global diametr
        self.name = name
        self.x = x
        self.y = y
        self.diametr = diametr
        if shouldAppend:
            circles.append(self)
            self.drawCircle()
    
    def drawCircle(self):
        canvas.create_oval(self.x, self.y, self.x + self.diametr, self.y + self.diametr, outline="black", fill="black", width=2)
    
def initCircles():
    global circles
    circles = []
    circle = cicrle(72,360,"A")
    circle = cicrle(181,517,"B")
    circle = cicrle(296,72,"C")
    circle = cicrle(399,617,"D")
    circle = cicrle(480,440,"E")
    circle = cicrle(617,140,"F")
    
    
def getCicrcleByPoint(point):
    global circles
    for circle in circles:
        if point.x > circle.x and point.x < circle.x + circle.diametr and point.y > circle.y and point.y < circle.y + circle.diametr:
            return circle
        
def getCircleByName(name):
    global circles
    for circle in circles:
        if circle.name == name:
            return circle
```

```python
# ⚠️⚠️⚠️⚠️ Points

class point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
startCircle = cicrle(0,0,"", False)
endCircle = cicrle(0,0,"", False)
```

```python
# ⚠️⚠️⚠️⚠️ Lines

lines = []
class line:
    def __init__(self, startCircle, endCircle):
        self.startCircle = startCircle
        self.endCircle = endCircle

def initLines():
    global lines
    for line in lines:
        canvas.create_line(line.startCircle.x + diametr / 2, line.startCircle.y + diametr / 2, line.endCircle.x + diametr / 2, line.endCircle.y + diametr / 2)
        
        
```

```python
# ⚠️⚠️⚠️⚠️ Mouse actions
isDrawing = False

def press(event):
    global startCircle
    global isDrawing 
    
    startCircle = getCicrcleByPoint(point(event.x, event.y))
    if (startCircle):
        isDrawing = True

def onMove(event):
    global isDrawing
    global startCircle

    if startCircle and isDrawing:
        refreshCanvas()
        canvas.create_line(startCircle.x + diametr / 2, startCircle.y + diametr / 2, event.x, event.y)
    
def release(event):
    global startCircle
    global endCircle
    global isDrawing 
    
    endCircle = getCicrcleByPoint(point(event.x, event.y))
    if endCircle:
        currentLine = line(startCircle,endCircle)
        isLine = isLineInTree(currentLine)

        if isLine and isDrawing:
            global lines
            lines.append(currentLine)
            computerStarts(currentTree)
 
    refreshCanvas()
    isDrawing = False

def refreshCanvas():
    canvas.delete("all")
    initCircles()
    initLines()
    
```

```python
# ⚠️⚠️⚠️⚠️ Popup message
import tkinter.messagebox

whoWins = -1

def showPopUp():
    global whoWins
    MsgBox = tk.messagebox.askquestion ('Izvele','Vai dators sāk spēli?',icon = 'warning')
    if MsgBox == 'yes':
        whoWins = 1
        computerStarts(currentTree)
    else:
        whoWins = -1
        
```

```python
# ⚠️⚠️⚠️⚠️ Copmuter turn

import random

def computerStarts(tree):
    global lines
    global currentTree
    
    try:
        iter(tree)
        winingBranches = []
        
        for branch in tree:
            if (branch.value == whoWins):
                winingBranches.append(branch)
        if len(winingBranches) != 0:
            randomWining = random.randint(0,len(winingBranches) - 1)
            branch = winingBranches[randomWining]
            currentLine = line(branch.startPoint,branch.endPoint)
            lines.append(currentLine)
            currentTree = branch.child
            if len(currentTree) == 0:
                showYouLose()
            refreshCanvas()
        else:
            showYouWin()
                  
    except TypeError:
        branch = tree
        currentLine = line(branch.startPoint,branch.endPoint)
        lines.append(currentLine)
        currentTree = branch.child
        showYouLose()
        
def showYouWin():
    global resultLabel
    resultLabel = tk.Label(root, text='Jūs uzvarejat!', fg='green', bg='white', font=("Arial", 40))
    resultLabel.place(relx = 0.5,
                rely = 0.5,
                anchor = 'center')

def showYouLose():
    global resultLabel
    resultLabel = tk.Label(root, text='Jūs zaudejat!', fg='red', bg='white', font=("Arial", 40))
    resultLabel.place(relx = 0.5,
                rely = 0.5,
                anchor = 'center')
    
```

```python
# ⚠️⚠️⚠️⚠️ Main

root = tk.Tk()
canvas = tk.Canvas(root, width=800, height=800)


def initAll():
    global lines
    global resultLabel
    lines = []
    
    try: 
        resultLabel.destroy()
    except NameError:
        print()
    
    
    initCircles()
    setTree()
    showPopUp()
    
    btn = tk.Button(root, text='Sākt no jauna', width=20,
             height=2, bd='10', command=initAll)
    btn.place(relx = 0.5,
              rely = 0.95,
              anchor = 'center')
    
    root.bind("<Button-1>", press)
    root.bind('<Motion>', onMove)
    root.bind("<ButtonRelease-1>",release)
    
def main():
    
    canvas.pack()
    
    initAll()
    
    root.mainloop()

main()
```
